## Imóveis Residenciais

```python
# Importando Pandas
import pandas as pd
```

```python
# Pegando dados do csv
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
# Mostrando 10 primeiras linhas do DataFrame
dados.head(10)
```

```python
# Pegando a coluna 'Tipo', tirando os resultados duplicados e transformando em uma lista
list(dados['Tipo'].drop_duplicates())
```

```python
# Deixando na lista somente os itens que queremos da lista anterior
residencial = ['Quitinete',
               'Casa',
               'Apartamento',
               'Casa de Condomínio',
               'Casa de Vila']
```

```python
# Mostrando o resultado
residencial
```

```python
# Mostrando novamente as 10 primeiras linhas de dados
dados.head(10)
```

```python
# Usando o 'isin' para bater a series 'dados['Tipo']' com a lista 'residencial', retornando True pro 
# que consta nos dois e False quando não consta
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# Criando um novo DataFrame, usando o 'dados' de template com a nova Series 'selecao' como parametro.
# Isso faz com que dados_residencial apresente só as linhas onde a coluna 'Tipo' dele for 'True' 
# na series 'selecao'
dados_residencial = dados[selecao]
```

```python
# Resultado
dados_residencial
```

```python
# Pegando a coluna 'Tipo' e tirando os duplicados pra ver se deu certo a selecao
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# Quantas linhas restaram em dados_residencial
dados_residencial.shape[0]
```

```python
# Quantas linhas tinham em dados
dados.shape[0]
```

```python
# Reformulando o index usando o range com o numero atual de linhas
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
# Resultado
dados_residencial
```

## Exercício

Veja o código que cria um DataFrame de letras e números.

```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df2 = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

```python
df2
```

E baseado nesses dataframe geramos uma selecao para filtrar os dados:

```python
selecao = df2['N'].isin([i for i in range(11) if i % 2 == 0])
df2 = df2[selecao]
df2
```
