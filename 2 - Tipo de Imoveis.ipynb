```python
# DataFrame do arquivo aluguel.csv
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
# Armazenando a Series em uma variável
tipo_de_imovel = dados['Tipo']
```

```python
# Mostrando que é uma series
type(tipo_de_imovel)
```

```python
# Faz uma previsão de como seria se tirar as duplicadas da series
tipo_de_imovel.drop_duplicates()
```

```python
# Mostrando que as duplicada não foram embora de verdade
tipo_de_imovel
```

```python
# Mandando as duplicadas embora de verdade
tipo_de_imovel.drop_duplicates(inplace = True)
```

```python
# Mostrando que agora é sério
tipo_de_imovel
```

```python
# Vendo se a mudança na Series muda o DataFrame inteiro. Resultado da análise: Não muda
dados
```

## Exercício

O pandas disponibiliza um método de DataFrame para remoção de linhas duplicadas. Este método é o drop_duplicates(), que apresenta um conjunto de argumentos de configuração. Considerando as duas linhas de código abaixo:

```python
import pandas as pd
dados2 = pd.DataFrame([1, 1, 2, 3, 3, 3, 4, 4], columns = ['X'])
dados2
```

Assinale a opção onde o método drop_duplicates() é corretamente aplicado!

#### Resposta:

```python
dados2.drop_duplicates(inplace = True)
dados2
```
