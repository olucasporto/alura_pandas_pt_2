## Tratamento de Dados Faltantes

```python
# Importando Pandas
import pandas as pd
```

```python
# Pegando dados de aluguel_residencial.csv
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=";")
```

```python
# Mostrando as 10 primeiras linhas do resultado
dados.head(10)
```

```python
# DataFrame mostrando quais dados são nulos (True) e quais não são (False)
dados.isnull()
```

```python
# Fazendo o Oposto do anterior
dados.notnull()
```

```python
# Usando info para saber quantos dados não nulos temos em relação aos nulos
dados.info()
```

```python
# Pegando a coluna 'Valor' e vendo quais dados são nulos
dados['Valor'].isnull()
```

```python
# Usando o resultado anterior para filtrar as linhas com os dados nulos em 'Valor'
dados[dados['Valor'].isnull()]
```

```python
# Quantidade de linhas no DataFrame original antes de tirar os valores nulos
A = dados.shape[0]
# Tirando os valores nulos
dados.dropna(subset = ['Valor'], inplace = True)
# Quantidade de linhas resultantes
B = dados.shape[0]
# Quantidade de linhas que foram embora
A - B
```

```python
# Resultado do dataframe com valores nulos em 'Valor' depois de dropas as linhas
dados[dados['Valor'].isnull()]
```

## Tratamento de dados faltantes (continuação)

```python
# Vendo linhas com valores nulos na coluna 'Condominio'
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Vendo quais valores na coluna 'Tipo' são 'Apartamento' e ao mesmo tempo tem valor nulo na coluna 
# 'Condominio'
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# Quantos dados tinha antes da seleção
A = dados.shape[0]
# Fazendo a seleção, mas com o "~", vai selecionar o oposto (selecao anterior vai sair do DataFrame)
dados = dados[~selecao]
# Quantos dados tem depois da seleção
B = dados.shape[0]
# Quantos dados foram retirados
A - B
```

```python
# Vendo quantos Condominios com valores nulos sobraram
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Preenchendo os valores nulos com 0 em 'Condominio' e 'IPTU'
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
# Quantos valores nulos sobraram em 'Condominio'
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Quantos valores nulos sobraram em 'IPTU'
dados[dados['IPTU'].isnull()].shape[0]
```

```python
# Mostrando que não existem mais valores nulos no DataFrame
dados.info()
```

```python
# Exportando para o arquivo csv
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
